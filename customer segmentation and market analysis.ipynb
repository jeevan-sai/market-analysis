### error detection

```python
for i in cols:
    skew=df[i].skew().round()
##meanstd()
    if skew == 0:
        m=df[i].mean()
        s=df[i].std()
        up=m+3*s
        lp=m-3*s
        df[i]=df[i].map(lambda x : up if x>up else lp if x<lp else x)
##iqr
    else :
         q1=df[i].quantile(0.25)
         q3=df[i].quantile(0.75)
         iqr=q3-q1
         upper=q3+1.5*iqr
         lower=q1-1.5*iqr
         df[i]=df[i].map(lambda x:  upper if x>upper  else lower if x<lower else x)
```
